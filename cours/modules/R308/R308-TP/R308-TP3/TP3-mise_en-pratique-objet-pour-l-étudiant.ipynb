````python
from io import StringIO

class TexteIter:
  def __init__(self, lines: list[str]) -> None:
    self.content = lines
    self.current = 0
    self.previous = 0
    self.end = False

  def __next__(self):
    if self.current+2 < len(self.content):
      self.previous = self.current
      self.current += 2

      return self.content[self.previous]
    else:
      if not self.end:
        self.end = True
        return self.content[-1]
      else:
        raise StopIteration

class Texte:
  """Cette classe renvoie une ligne sur deux d'un texte (pour décoder le sens caché)."""

  def __init__(self, text: str) -> None:
    self.body = StringIO(text)
    self._iterator_content = self.body.readlines()
    self._iterator_count = 0

  def __len__(self) -> int:
    return len(self._iterator_content)

  def __repr__(self) -> str:
    return self.body.getvalue()

  def __iter__(self) -> TexteIter:
    return TexteIter(self._iterator_content)

    
class Test:
  def __init__(self):
    self.count = 0
    self.max = 10

  def __iter__(self):
    return self
  
  def __next__(self):
    self.count += 2

    if self.count < self.max:
      return self.count

    raise StopIteration
````
