- Retrouvez une valeur avec D.get(clef)

````python
my_dict = dict()

appreciation = (
  ("windows", "bien mais si tu connais pas ce qu'est un PC."),
  ("linux", "Super bien mais si t'as plus d'ordi, c'est ton problème."),
  ("mac", "Dis-moi que t'as trop d'argent sans me le dire."),
  ("vms", "Personnellement, jamais utilisé."),
  ("aix", "C'est d'IBM ?"),
  ("solaris", "C'est d'Oracle, ça je connais.")
)

for elem in appreciation:
  key, value = elem
  my_dict[key] = value

print("Qu'est-ce que j'ai dis sur Windows déjà ?")
print(my_dict.get("windows"), end="\n\n")

print("D'accord mais maintenant pour le tout:")
for key, value in my_dict.items():
  print(f"\n- {key}: {value}")

print("\n") # Just to make some space

print("Ok, mais une chose à la fois ?")
print("D'abord, les clés:", end="\n\n")
curr_index = 0
for key in my_dict.keys():
  curr_index += 1
  print(f"Clé n°{curr_index}: {key}")

print("\n") # Just to make some space

print("Et maintenant les valeurs correspondantes:", end="\n\n")
curr_index = 0
for value in my_dict.values():
  curr_index += 1
  print(f"Valeur n°{curr_index}: {value}")

print("\n") # Just to make some space
````

````python
l_init = [["un", "deux", "trois"], 1, 2, 3]
l1 = l_init[:]
l2 = l1 # A symlinked copy

println("L1 est égal (en pointeurs) à L2" if l1 is l2 else "L1 n'est pas égal (en pointeurs) à L2")

print("J'enlève maintenant un 'un' à l1")
l1[0].pop(0)
println(f"l1: {l1}, l2: {l2}")
print("L1 est égal (en pointeurs) à L2" if l1 is l2 else "L1 n'est pas égal (en pointeurs) à L2")
````

````python
from copy import deepcopy, copy

l1 = [["un", "deux", "trois"], 1, 2, 3]

scp = copy(l1)
dcp = deepcopy(l1)

println("C'est une shallow copy de l1:", scp)
println("C'est une deep copy de l1:", dcp)

### Premier test sans enlever quoi que ce soit
print("La shallow copy est égal à L1" if scp == l1 else "La shallow copy n'est pas égal à L1")
println("La deep copy est égal à L1" if dcp == l1 else "La deep copy n'est pas égal à L1")

### Dexuième test en enlevant l'entrée [0][0]
l1[0].pop(0)
print("La shallow copy est égal à L1" if scp == l1 else "La shallow copy n'est pas égal à L1")
println("La deep copy est égal à L1" if dcp == l1 else "La deep copy n'est pas égal à L1")

println("On voit bien que la deep copy ne prend pas en compte les changements de la liste précédente:")
print(f"L1: {l1}, les copies: \n\t- {scp}\n\t- {dcp}")
````

## Exercice 4 Reformatage d'un fichier DHCP (source sous licence http://creativecommons.org/licenses/by-sa/3.0/)


Il vous est proposé dans l'exercice suivant de "parser" un fichier de configuration permettant la distribution d'adresses IP à des postes de travail. Dans un premier temps vous n'utiliserez pas la programmation orientée objet puis dans un second temps vous re-créerez votre code ens utilisant des objets.
 

Durée estimée 30 mn 6pts

Comme le formatage du fichier dhcpd.conf est très libre par rapport aux sauts de lignes, on va reformater le fichier à notre façon. Pour cela, on va fusionner toutes les lignes dans un string, que l'on pourra ensuite parser.

Il s'agit donc ici d'implémenter cet algorithme en Python:

    de lire le fichier ligne par ligne (utiliser le package fileinput)
    d'éliminer les lignes vides et les commentaires (vous pouvez utilisez utiliser re.sub(...))
    de concaténer le tout dans un string,
    d'extraire les portions host {…} de ce string, (utilisez re.findall(...))
    de stocker chaque portion extraite dans une liste hosts.
    de faire un dump de cette liste.

Extrait du Dump:
['host coquil-gw1 { hardware ethernet 00:1d:09:0a:33:db; fixed-address coquil-gw1; }', 'host perso-ducrot { hardware ethernet 00:18:8B:D2:1C:4E; fixed-address perso-ducrot; }', 'host fpl { hardware ethernet 00:26:b9:5d:32:32;...


````python
import fileinput
import re

dhcpd_conf = ""
with fileinput.input(files=("./src/dhcpd.conf"), encoding="utf-8") as fin:
  for line in fin:
    if line == "\n":
      line = re.sub(r"\n", "", line)
    elif "#" in line:
      line = re.sub(r"^#*([a-zA-Z]*)+.+$\n", "", line)

    dhcpd_conf += line

print("Je récupère maintenant que les portions `host`")

hosts = [host[0] for host in re.findall(r"(host (\w[a-zA-Z]*-*[a-zA-Z]*) {\s.*})", dhcpd_conf)]

with open("./out/dhcpd_hosts.txt", "wt", encoding="utf-8") as fout:
  for host in hosts:
    fout.write(host + "\n")

print(f"In total, there is: {len(hosts)} hosts retrieved!")

````

````python
class Host:
    """ Classe représentant une machine"""
    def __init__(self, hostname:str, attributes):
        self.hostname = hostname
        self.attributes = attributes
          
    def __repr__(self):
        return f"DHCP::Host(hostname={self.hostname} attributes={self.attributes}"
````

````python
import re

class Database:
    """Classe relative au DHCP """
    def __init__(self):
        self.pattern={
            "host": "(\w[a-zA-Z]*-*[a-zA-Z]*)"
        }
        self.hosts=[]

    def addPattern(self, key: str, pattern: str):
        """ Méthode d'ajout d'une mac ou d'une IP """
        self.pattern[key] = pattern

    def dump(self):
        """Affichage de l'entrée hôte"""
        for host in self.hosts:
            print(host)

    def load(self,filename):
        """Chargement du fichier DHCP"""
        with open(filename, "rt", encoding="utf-8") as fin:
            if fin.readable():
                rfin = fin.readlines()
                for line in rfin:
                    curr_attributes = {
                        "ip": "",
                        "mac": ""
                    }

                    host = re.findall(self.pattern["host"], line)[1]
                    mac = re.findall(self.pattern["mac"], line)[0].strip().replace(";", "")
                    ip = re.findall(self.pattern["ip"], line)

                    ### Let's set the attributes dict
                    curr_attributes.update({
                        "ip": "".join(ip).strip().replace(";", ""),
                        "mac": mac
                    })

                    self.hosts.append(Host(host, curr_attributes))
````

````python

chemin="./out/dhcpd_hosts.txt"
db = Database()
db.addPattern( "mac", 'hardware\s+ethernet\s+(\S+)' )
db.addPattern( "ip",  'fixed-address\s+(\S+)'       )
db.load(chemin)
db.dump()
````
