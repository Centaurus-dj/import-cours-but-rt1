```python
import pandas as pd
import matplotlib.pyplot as plt

seq = pd.Series([43, 52])

x_labels = [2018, 2021]

plt.figure(figsize=(12, 12))
fig = seq.plot(kind="bar")
fig.set_title("Percentage of people feeling burn-out or having experienced a burn-out in 2019 and 2022")
fig.set_xlabel("Year")
fig.set_ylabel("Feeling of being burned-out or having experienced a burn-out (in %)")
fig.set_xticklabels(x_labels)
```
