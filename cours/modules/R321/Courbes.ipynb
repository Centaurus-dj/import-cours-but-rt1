```python
import numpy as np
import matplotlib.pylab as plt


F = 10
T = np.linspace(0,0.3,1001)
pi = np.pi

et = np.sin(2*pi*F*T)
st = 0.708 * np.sin(2*pi*F*T  + (- pi / 4))

plt.plot(T, et, label="e(t)", color="green")
plt.plot(T, st, label="s(t)", color="red")
plt.xlabel('Temps (ms)')
plt.ylabel('Amplitude')
plt.axis('tight')
plt.legend()
plt.show()
```
